```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

weights = [0.859, 0.833]

data = pd.read_csv("./vaxData.csv", thousands=",", parse_dates=["date"])

for i in data[data["date"] < pd.to_datetime("2021-07-02")].index:
    data.at[i, "first"] = data.at[i, "aged"] + data.at[i, "primary"] + data.at[i, "state"]

data = pd.read_csv("./vaxData.csv", thousands=",", parse_dates=["date"])
for i in data[(pd.notna(data["second"])) & (data["date"] < pd.to_datetime("2021-07-29"))].index:
    data.at[i, "first"] = data.at[i, "first"] * weights[0]
    data.at[i, "second"] = data.at[i, "second"] * weights[1]
    
data.plot("date")
    
# data = data[["date", "first", "second"]].dropna()
# data_a = data1[data1["date"] > pd.to_datetime("2021-07-28")]
# data_b = data1[data1["date"] < pd.to_datetime("2021-07-29")]

# X = data_a.iloc[:, 1].values.reshape(-1, 1)
# Y = data_a.iloc[:, 2].values.reshape(-1, 1)
# regressor = LinearRegression()
# regressor.fit(X, Y)
# pred = regressor.predict(X)

# plt.scatter(X, Y)
# plt.plot(X, pred, color='red')
# plt.show()
```
