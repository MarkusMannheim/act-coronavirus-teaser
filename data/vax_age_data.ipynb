```python
import pandas as pd, datetime as dt
import time

vax_data = pd.DataFrame(columns=["date"]).set_index("date")
date = dt.date.today()

while date > dt.date(2021, 9, 4):
    
    print(f"checking data for {date:%b %#d, %Y} ...", end="\r")
    error = False
    
    urls = [
        f"{date:%#d-%B-%Y}",
        f"{date:%#d-%B-%Y}_0",
        f"{date:%#d-%B-%Y}_1",
        f"{date:%#d-%B-%Y}_2",
        f"{date:%#d-%B-%Y}-covid-19-vaccination-vaccination-data-{date - dt.timedelta(days=1):%#d-%B-%Y}",
        f"{date:%#d-%B-%Y}-covid-19-vaccination-vaccination-data-{date:%#d-%B-%Y}"        
    ]
    
    for i, url in enumerate(urls):
        try:
            data = pd.read_excel(
                f"https://www.health.gov.au/sites/default/files/documents/{date:%Y}/{date:%m}/covid-19-vaccination-vaccination-data-{url}.xlsx".lower(),
                na_values=["-"],
                names=["measure", "value"],
                index_col="measure"
            )
            break
        
        except:
            if i == len(urls) - 1:                
                print(f"checking data for {date:%b %#d, %Y} ... inaccessible!")
                error = True
    
    if not error:
        print(f"checking data for {date:%b %#d, %Y} ... accessible")
        
        data = data.loc[[x for x in data.index[pd.notna(data.index)] if "ACT - " in x]]
        
        for m in data.index:
            vaxData.at[date, m] = data.at[m, "value"]
        
    date = date - dt.timedelta(days=1)

vaxData
```
