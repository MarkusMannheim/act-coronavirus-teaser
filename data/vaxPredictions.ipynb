```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns, matplotlib.ticker as ticker, matplotlib.dates as mdates, statsmodels.api as sm, statsmodels.formula.api as smf
import io, requests
from PyPDF2 import PdfFileReader
sns.set(style="darkgrid", font_scale=1.25)

vaxData = pd.DataFrame(columns=["date", "first", "second"])
vaxData.set_index("date", inplace=True)
population = 431826 - 27332 - 29006 - 5595 - 5513
    
def vaxScrape(date):
    print(f"checking vaccination data for {date:%B %#d, %Y} ...")
    url = f"https://www.health.gov.au/sites/default/files/documents/{date:%Y}/{date:%m}/covid-19-vaccine-rollout-update-{date:%#d-%B-%Y}.pdf".lower()

    try:
        request = requests.get(url)
        file = io.BytesIO(request.content)
        reader = PdfFileReader(file)
        contents = reader.getPage(5).extractText().split("\n")

        try:
            first_16 = float(contents[contents.index("dose 1") + 2].strip().replace(",", ""))
            first_12 = float(contents[len(contents) - contents[::-1].index("dose 1") + 1].strip().replace(",", ""))
            second_16 = float(contents[contents.index("dose 2") + 2].strip().replace(",", ""))
            second_12 = float(contents[len(contents) - contents[::-1].index("dose 2") + 1].strip().replace(",", ""))
            vaxData.at[date, "first"] = (first_16 + first_12)
            vaxData.at[date, "second"] = (second_16 + second_12)
            
        except:
            print("couldn't read data for this date")            
            
    except:
        print("couldn't access this data")

today = pd.to_datetime(pd.Timestamp("now").date())
i = 0
while i < 28:
    vaxScrape(today - pd.Timedelta(days=i))
    i = i + 1

vaxData["day"] = vaxData.index.dayofyear
vaxData.sort_values("day", inplace=True)
vaxData["first"] = pd.to_numeric(vaxData["first"])
vaxData["second"] = pd.to_numeric(vaxData["second"])

i = 0
while i < 21:
    previousDate = vaxData.index[-1]
    previousDay = vaxData.at[vaxData.index[-1], "day"]
    date = previousDate + pd.Timedelta(days=1)
    day = previousDay + 1
    vaxData.at[date, "day"] = day
    i = i + 1

for date in vaxData.index:
    first = vaxData.at[date, "first"]
    second = vaxData.at[date, "second"]
    day = vaxData.at[date, "day"]
    previousDate = date - pd.Timedelta(days=1)
    if previousDate in vaxData.index:
        vaxData.at[date, "first_shots"] = first - vaxData.at[previousDate, "first"]
        vaxData.at[date, "second_shots"] = second - vaxData.at[previousDate, "second"]

result = smf.ols("second_shots ~ day + " + "I(1/day)", data=vaxData.dropna()).fit()
vaxData["predicted"] = result.predict(vaxData["day"])

for i in vaxData.index:
    if not np.isnan(vaxData.at[i, "first"]):
        vaxData.at[i, "predicted"] = np.nan

fig, ax = plt.subplots(
    tight_layout=True,
    figsize=(9, 6)
)

sns.scatterplot(x=vaxData.index, y=vaxData["second_shots"], ax=ax, s=100, label="observed")
sns.scatterplot(x=vaxData.index, y=vaxData["predicted"], ax=ax, s=100, label="projected")
plt.ylabel("second doses")
plt.xlabel(None)
plt.title("Projected daily vaccinations (ACT)", fontweight="bold", fontsize=18)
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %#d"))
plt.savefig("vaxProjection.jpg")

print()
print("target doses:", population * .99)
print() 
vaxData["predicted"] = vaxData["second"].max() + vaxData["predicted"].cumsum()
vaxData
```

```python
result_linear = smf.ols("second ~ day", data=vaxData).fit()
result_poly = smf.ols("second ~ day + " + "I(day**2)", data=vaxData).fit()
result_log = smf.ols("second ~ day + " + "I(1/day)", data=vaxData).fit()

fig, ax = plt.subplots(
    tight_layout=True,
    figsize=(12, 8)
)
sns.scatterplot(x=vaxData.day, y=vaxData.second, ax=ax, s=100)
sns.lineplot(x=vaxData["day"], y=result_linear.predict(vaxData["day"]), ax=ax)
sns.lineplot(x=vaxData["day"], y=result_poly.predict(vaxData["day"]), ax=ax)
sns.lineplot(x=vaxData["day"], y=result_log.predict(vaxData["day"]), ax=ax)

print("linear:", result_linear.rsquared, result_linear.pvalues[1])
print("poynomial:", result_poly.rsquared, result_poly.pvalues[1])
print("log:", result_log.rsquared, result_log.pvalues[1])

# weights = np.polyfit(x, y, 2)
# model = np.poly1d(weights)
# results = smf.ols(formula="y ~ model(x)", data=df).fit()
# results.summary()
# results.predict(x)
# sns.lineplot(data=df, x="x", y="y", ax=ax, label="first dose", lw=3)
# sns.lineplot(data=vaxData, x="x", y="second", ax=ax, label="second dose", lw=3)
# ax.yaxis.set_major_formatter(ticker.PercentFormatter(1, 0))
# plt.xlabel("day of year")
# plt.ylabel("eligible population (12 years +)")
```
